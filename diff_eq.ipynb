```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
h = 1/(n-1)
axis = [0]*n
y_analytic = [0]*n
y_approx = [1]*n
error = [0]*n

for i in range(n):
    axis[i] = i/(n - 1)
    if i > 0: 
        y_approx[i] = (1 - h)*y_approx[i - 1]
    y_analytic[i] = np.exp(-axis[i])
    error[i] = np.abs(y_analytic[i] - y_approx[i])/h
    
plt.plot(axis, y_approx, label = 'approx')
plt.plot(axis, y_analytic, label = 'analytics')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(axis, error, label = 'error')
plt.legend()
plt.grid(True)
plt.show()
```

```python
Num = [5, 10, 20, 40, 60, 80, 100, 150, 200, 400, 500, 1000]
H = [0]*len(Num)
bound = [0.205, 0.193, 0.188, 0.18593, 0.18525, 0.1849, 0.1847, 0.1845, 0.1843, 0.1841, 0.18409, 0.18402]

for i in range(len(Num)):
    H[i] = 1/(Num[i] - 1)
    bound[i] /= H[i]

plt.plot(H, bound)
plt.show()
```

```python
log_bound = [0]*len(Num)
log_H = [0]*len(Num)

for i in range(len(Num)):
    log_bound[i] = np.log(bound[i])
    log_H[i] = np.abs(np.log(H[i]))
p = (log_bound[len(Num) - 1] - log_bound[2])/(log_H[len(Num) - 1] - log_H[2]) 
    
plt.plot(log_H, log_bound)
plt.show()
```

```python
n = 10
h = 1/(n-1)
axis = [0]*n
y_analytic = [0]*n
y_approx_new = [1]*n
error = [0]*n

for i in range(n):
    axis[i] = i/(n - 1)
    if i > 0: 
        y_approx_new[i] = y_approx_new[i - 1]/(1 + h)
    y_analytic[i] = np.exp(-axis[i])
    error[i] = np.abs(y_analytic[i] - y_approx_new[i])/h
    
plt.plot(axis, y_approx_new, label = 'approx')
plt.plot(axis, y_analytic, label = 'analytics')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(axis, error, label = 'error')
plt.legend()
plt.grid(True)
plt.show()
```
